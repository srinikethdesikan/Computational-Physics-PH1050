### Problem 1
Solve the first order ordinary differential equation (ODE) xy â€² + y = 0 using the Euler method with an initial
condition y (x = 1) = 5. Plot both the analytical and numerical solution and their difference from x = 1 to
x = 5 with 100 steps.

```python
x=np.linspace(1,5,100)
y=np.zeros(len(x))
y[np.where(x==1)[0]]=5 #Initial condition
def y1(x,y):
    return (-y/x)
h=x[1]-x[0]
for i in range(0,len(x)-1):
    y[i+1]=y[i]+h*y1(x[i],y[i])
print(y)
plt.plot(x,y,linestyle='--',label='Numeric Solution')
# As we know, the analytic solution is y=5/x
plt.plot(x,5/x,linestyle=':',label='Analytic Solution')
plt.title("Solution to the DE")
plt.legend()
plt.show()
plt.plot(x,abs(y-5/x),label='Absolute error Plot')
plt.title("Error Plot")
plt.legend()
plt.show()
```

### Problem 2
Solve the the differential equation $ m\ddot{x}+ kx = 0$ numerically using Euler method with an initial conditions
x(t = 0) = 0 and v (t = 0) = 5m/s. Assume m=k=1. Plot the x(t), v(t) for a time range t = 0 to t = 20.
Also, calculate the total energy (E) in every time step, and plot E vs t.

```python
# The above question can be solved by treating it as 2 differential equations
t=np.linspace(0,20,100000)
x=np.zeros(len(t))
v=np.zeros(len(t))
E=np.zeros(len(t))
def v1(x,k=1,m=1):
    return -k*x*(1/m)
def x1(v):
    return v
h=t[1]-t[0]
x[0]=0
v[0]=5
for i in range(0,len(t)-1):
    x[i+1]=x[i]+h*x1(v[i])
    v[i+1]=v[i]+h*v1(x[i])
E=(0.5)*(x**2+v**2)
U=0.5*(x**2)
T=0.5*(v**2)
plt.style.use('default')
plt.plot(t,x,color='green',linestyle='-',label='Position')
plt.plot(t,v,color='blue',linestyle=':',label='Velocity')
plt.title('Position and Velocity')
plt.xlabel("Time (t)")
plt.ylabel("x(t), v(t)")
plt.legend()
plt.show()
plt.plot(t,E,label='Total Energy')
plt.plot(t,U,label='Potential Energy')
plt.plot(t,T,label='Kinetic energy')
plt.title("Energy")
plt.xlabel('Time')
plt.ylabel('Energy')
plt.legend()
plt.show()
```

Note that the accuracy becomes better as the no. of t values sampled increases

### Problem 3
Define a function Eul(f(x,y), x0, y0, h, N) that will return the value of y after N steps of size h. Use this to solve the two problems below:
1. Solve the ODE $(1 + x)
\frac{dy}{dx} âˆ’ 2y + 18x = 0$ with y(0)=4 with h=0.05 in the interval (0, 3) (i.e. 60 steps) by using Eulerâ€™s method.
2. Plot the analytical solution $(y = âˆ’5x^2 + 8x + 4)$ and the numerical solution in the same figure to visually inspect the outcome. Plot the absolute error at each point for Eulerâ€™s method as a
function of x.    

```python
def Eul(y1, x0, y0, h, N):
  # y'=(y(x+h)-y(x))/h
  x=x0
  y=y0
  for i in range(N):
    y+=h*y1(x,y)
    x+=h
  return y
```

```python
def y1(x,y):
  return (2*y-18*x)/(1+x)

l=[]
for n in range(60):
  l.append(Eul(y1, 0, 4, 0.05, n))
print("The value of y when x is 3 is",Eul(y1, 0, 4, 0.05, 60))
print(l)
x=np.linspace(0,3,60)
plt.plot(x,l,color='b')
plt.plot(x,-5*x**2+8*x+4,color='r')
plt.title('Solutions of the equation')
plt.legend(['Numerical','Analytical'])
plt.show()
l=np.array(l)
y=-5*x**2+8*x+4
e=abs(y-l)
plt.plot(x,e)
plt.title('Absolute error as a function of x')
plt.show()
```

### Problem 5
**Coupled ODE** While solving a set of M coupled ODEs the function â€˜Eulâ€™ may be extended to return an array
Y= Eul (F(x,y), x0, y0, h, N,M ), where Y is the M-dimensional array of solutions and F is a M-dimensional
array of RHS functions for coupled ODEs. Using this modified function find the phase space trajectory of a
particle of unit mass in a potential V(x).
(a) Start by taking simple harmonic oscillator potential: $V (x) = \frac{1}{2}kx^2$
, k = 1N/m. What is the period T
of this oscillator? Use h = T /S with S=10, 100 and 1000 for the step size and follow the trajectory over a
time 10 T considering the initial condition (x, px ) = (1.0, 0.0). Fro all the cases, compare the following
(i) Show the trajectory i.e. x and p as functions of time and also the phase space . Compare these with the
analytical solutions.
(ii) The energy E is expected to be conserved. Plot the relative change in energy dE/E = (En âˆ’ E)/E as
a function of time, here En and E are respectively the numerically calculated energy and the actual energy
respectively.

```python
def EulC(F, x0, yx0, h, N, M):
    '''
    F is a list of derivative functions of 1 independent variable and M dependent variables.
    x0 is the initial value of the independent variable x.
    yx0 is an array of initial conditions for the dependent variables.
    h is the step size.
    N is the number of steps.
    '''
    x = np.linspace(x0, x0 + h * N, N)
    u = np.zeros((M, N))
    for i in range(M):
        u[i][0] = yx0[i]
    for j in range(N - 1):
        for i in range(M):
            u[i][j + 1] = u[i][j] + (h / 2) * (x[j])
    
    return x, u
```

```python
def dxdt(t, x, p):
    return p

def dpdt(t, x, p):
    return -x

F = [dxdt, dpdt]
x0 = 0
y0 = [1.0, 0.0] 
N = 1000 
T = 2 * np.pi 
S = [10, 100, 1000]  # Different step sizes

for s in S:
    h = T / s
    time, Y = EulC(F, x0, y0, h, int(10 * T / h), 2)
    
    # Plot trajectory x(t) and p(t)
    plt.plot(time, Y[0, :], label='x(t)')
    plt.plot(time, Y[1, :], label='p(t)')
    plt.legend()
    plt.title(f'Trajectory with S={s}')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.show()
    
    # Plot phase space trajectory
    plt.plot(Y[0, :], Y[1, :])
    plt.title(f'Phase Space with S={s}')
    plt.xlabel('x')
    plt.ylabel('p')
    plt.show()
    
    # Calculate and plot relative energy change
    E = 0.5 * (Y[1, :]**2 + Y[0, :]**2)  # Total energy
    dE = (E - E[0]) / E[0]
    
    plt.plot(time, dE)
    plt.title(f'Relative Change in Energy with S={s}')
    plt.xlabel('Time')
    plt.ylabel('dE/E')
    plt.show()
```

### Problem 7
Define a function â€˜RK2 =(f(x,y), x0, y0, h, N)â€™ which will return the solution of a ODE of the form dy/dx =f (x, y ) for N steps of size h using 2nd order Runge-Kutta (RK) method.
Solve the ODE (1+x)
dy/dx âˆ’2y + 18x = 0 with y(0)=4 interval h=0.05 in the interval (0, 3) by using function
RK2. Compare this with the analytical solution and that from Eulerâ€™s method.

```python
# denote dy/dx by y1
def RK2(y1,x0,y0,h,N):
    x=np.zeros(N)
    y=np.zeros(N)
    x[0]=x0
    y[0]=y0
    for i in range(0,N-1):
        k1=y1(x[i],y[i])
        k2=y1(x[i]+h,y[i]+h*k1)
        y[i+1]=y[i]+(h/2)*(k1+k2)
        x[i+1]=x[i]+h
    return x,y
        
```

```python
def g1(x,y):
    return (2*y-18*x)/(1+x)
x,yN=RK2(g1,0,4,0.05,int(3/0.05))
yE=[4]
for i in range(len(x)-1):
    yE.append(Eul(g1,x[i],yE[i],0.05,1))
plt.plot(x,yN,label='Numerical - RK2')
plt.plot(x,yE,label='Numerical - Euler')
plt.plot(x,(-5*x**2+8*x+4),label='Analytical')
plt.legend()
plt.show()
```

Note that the Euler method is less accurate than the RK method

### Problem 8
**Coupled ODE**: Extend the function â€˜RK2â€™ to generalize it for M coupled ODEs as Y= RK2 (F(x,y), x0, y0,
h, N,M ), where Y is the array of solutions and F is a M-dimensional array of RHS functions for coupled ODEs. Solve the following coupled ODE by R-K 2nd order method for x = [0.0,0.5] with h=0.05. Initial
conditions are y(0)=0, z(0)=1.
$ \frac{dy}{dx} = âˆ’x âˆ’ yz$
$ \frac{dz}{dx} = âˆ’y âˆ’ xz $
Plot the solutions y(x) and z(x).

```python
def RK2C(F, x0, yx0, h, N, M):
    '''
    F is a list of derivative functions of 1 independent variable and M dependent variables.
    x0 is the initial value of the independent variable x.
    yx0 is an array of initial conditions for the dependent variables.
    h is the step size.
    N is the number of steps.
    '''
    x = np.linspace(x0, x0 + h * N, N)
    u = np.zeros((M, N))
    for i in range(M):
        u[i][0] = yx0[i]
    for j in range(N - 1):
        for i in range(M):
            k1 = F[i](x[j], *u[:, j])
            k2 = F[i](x[j] + h, *u[:, j] + h * k1)
            u[i][j + 1] = u[i][j] + (h / 2) * (k1 + k2)
    
    return x, u
```

```python
def y1(x,y,z):
    return -x-(y*z)
def z1(x,y,z):
    return -y-(x*z)
x,Y=RK2C([y1,z1],0,[0,4],0.05,10,2)
m,n=np.shape(Y)
labels=['y','z']
for j in range(m):
    plt.plot(x,Y[j,:],label=str(labels[j])+'(x)')
plt.grid()
plt.legend()
plt.show()
```
