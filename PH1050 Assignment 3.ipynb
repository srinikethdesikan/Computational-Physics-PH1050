### Problem 5
Evaluation of Sin(x) within a given error limit by adding up the series: Use power series expansion in order
to evaluate Sin(x) for a given x. Truncate the series when the value reaches within the accuracy of allowed error e
(User defined error limit) set by you.

```python
# Let the error limit be specified by e:
def sin(x,e):
    s=0
    i=1
    j=0
    while 1:
        if abs((x**i)*((-1)**j)/mt.factorial(i))>e:
            s=s+(x**i)*((-1)**j)/mt.factorial(i)
            i=i+2
            j=j+1
        else:
            break
        
    return s
```

```python
print(sin(2,0.01))
```

```python
def logic(x,A):
    l1=range(0,100)
    for i in range(0,99):
        x.append(A*x[i]*(1-x[i]))
    plt.plot(l1,x)
    plt.show()
```
